### 2.2. Creating a Line Graph

```r
head(pressure)
```

```r
# Gráfico de linha nativo do R
plot(pressure$temperature, pressure$pressure, type = 'l')
```

```r
# Adicionando pontos
plot(pressure$temperature, pressure$pressure, type = 'l')
points(pressure$temperature, pressure$pressure)

lines(pressure$temperature, pressure$pressure/2, type = 'l', col='red')
points(pressure$temperature, pressure$pressure/2, col='red')
```

```r
# Equivalentemente no ggplot
qplot(pressure$temperature, pressure$pressure, geom = 'line')
```

```r
# Ou ainda
qplot(data = pressure, temperature, pressure, geom = 'line')
```

```r
# Ou por fim
ggplot(pressure, aes(temperature, pressure)) +
    geom_line()
```

```r
# Ou
qplot(data=pressure, temperature, pressure, geom = c('line', 'point'))
```

```r
ggplot(pressure, aes(temperature, pressure)) +
    geom_line()+
    geom_point()
```

```r
ggplot(pressure) +
    geom_line(aes(temperature, pressure), col='blue') +
    geom_point(aes(temperature, pressure), col='blue') +
    geom_line(aes(temperature, pressure/2), col='red') +
    geom_point(aes(temperature, pressure/2), col='red')
```

# Plotting a Function Curve

```r
curve(x^2, from = -4, to = 4)
```

```r
meu_plot <- function(x)
{
   1/(1 + exp(-x + 10))
}
```

```r
curve(meu_plot(x), from=0, to=20)
# Add a line:
curve(1-meu_plot(x), add = TRUE, col = "red")
```

```r
library(ggplot2)
```

```r
ggplot(data.frame(c(0, 20)), aes(x=x)) +
 stat_function(fun=meu_plot, geom = 'line')
```
